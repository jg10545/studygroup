# Let's do some machine learning

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

%matplotlib inline
from IPython.core.pylabtools import figsize
```

# Generate some fake data from a simple function

```python
N = 100
np.random.seed(1)
X_train = np.random.uniform(0, 2, N)
Y_train = np.sin(8*X_train)
```

```python
plt.plot(X_train, Y_train, ".")
plt.grid(True)
plt.xlabel("X", fontsize=14)
plt.ylabel("Y", fontsize=14);
```

# And try to approximate it with a completely overkill nonparametric machine learning model

```python
inpt = tf.keras.layers.Input((1))
net = inpt
for _ in range(3):
    net = tf.keras.layers.Dense(128, activation="relu")(net)

outpt = tf.keras.layers.Dense(1)(net)
model = tf.keras.Model(inpt, outpt)
model.compile(tf.keras.optimizers.Adam(),
             loss=tf.keras.losses.mse)
```

```python
model.fit(X_train.reshape(-1,1), Y_train, epochs=2000)
```

# Did it work?

```python
X_pred = np.linspace(0, 2, 500)
Y_pred = model.predict(X_pred)
```

```python
plt.plot(X_train, Y_train, "o", label="training points")
plt.plot(X_pred, Y_pred, label="predicted")
plt.legend(loc="lower right")
plt.xlabel("X", fontsize=14)
plt.ylabel("Y", fontsize=14)
plt.title("victory!")
plt.grid(True);
```

# Great! Now let's make a prediction a little outside the domain of the training data.

```python
X_deploy = np.linspace(2,3,100)
Y_deploy_gt = np.sin(8*X_deploy)
Y_deploy_pred = model.predict(X_deploy)
```

```python
plt.plot(X_train, Y_train, "o", label="training points")
plt.plot(np.concatenate([X_pred, X_deploy], 0), 
         np.concatenate([Y_pred, Y_deploy_pred],0), label="predicted")
plt.plot(X_deploy, Y_deploy_gt, "--", label="the answer we were looking for")
plt.legend(loc="lower left")
plt.xlabel("X", fontsize=14)
plt.ylabel("Y", fontsize=14)
plt.grid(True);
```
